### Assignment 2: Drug Effectiveness and Side Effects Comparison
Objective:
Compare drug effectiveness and side effects by different products, regions, and trial periods. Visualize these comparisons using matplotlib and seaborn.
Instructions:
Load the dataset.
Perform any necessary data cleaning.
Create the following visualizations:
A bar plot comparing the average Effectiveness for each drug across different regions.
A violin plot to show the distribution of Effectiveness and Side_Effects for each product.
A pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend.
A boxplot comparing Effectiveness for different trial periods.
A regression plot to analyze how Marketing_Spend affects drug Effectiveness.
Based on the visualizations, provide an analysis of:
Which product has the best overall effectiveness.
The correlation between effectiveness and side effects.
Expected Outcome:
Insights into how the effectiveness of each drug varies by region and trial period.
Understanding the trade-off between drug effectiveness and side effects.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("sales_data.csv")

# Data Cleaning: Drop missing values
df.dropna(inplace=True)

# Bar plot: Average Effectiveness by Drug across Regions
plt.figure(figsize=(10, 5))
sns.barplot(x='Product_Name', y='Effectiveness', hue='Region', data=df, estimator=lambda x: x.mean())
plt.title('Average Effectiveness of Drugs Across Regions')
plt.xticks(rotation=45)
plt.legend(title='Region')
plt.show()

# Violin plot: Distribution of Effectiveness and Side Effects per Product
plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, inner='quartile')
plt.title('Effectiveness Distribution by Drug')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, inner='quartile')
plt.title('Side Effects Distribution by Drug')
plt.xticks(rotation=45)
plt.show()

# Pairplot: Effectiveness, Side_Effects, and Marketing Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.show()

# Boxplot: Effectiveness for different trial periods
plt.figure(figsize=(10, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness by Trial Period')
plt.show()

# Regression plot: Marketing Spend vs Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df)
plt.title('Effectiveness vs Marketing Spend')
plt.show()

# Analysis
best_drug = df.groupby('Product_Name')['Effectiveness'].mean().idxmax()
corr_effect_side = df[['Effectiveness', 'Side_Effects']].corr().iloc[0, 1]

print(f"The most effective drug is {best_drug}.")
print(f"The correlation between effectiveness and side effects is {corr_effect_side:.2f}.")
```
