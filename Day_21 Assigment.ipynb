### Part 1: Data Exploration
Load the dataset and display the first few rows. What are the independent and dependent variables?
Check for missing values. How would you handle them if any exist?
Generate summary statistics (mean, median, standard deviation) for numerical variables.
Create a correlation heatmap and interpret the correlation between Construction Cost and other factors.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Civil_Engineering_Regression_Dataset.csv")

# Display first few rows
print("First few rows of the dataset:")
print(df.head())

# Identify independent and dependent variables
dependent_variable = "Construction_Cost"  # Assuming this is the dependent variable
independent_variables = [col for col in df.columns if col != dependent_variable]
print("\nDependent Variable:", dependent_variable)
print("Independent Variables:", independent_variables)

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing values in the dataset:")
print(missing_values[missing_values > 0])

# Handle missing values (if any exist)
df.fillna(df.median(), inplace=True)  # Filling missing values with the median

# Generate summary statistics
summary_statistics = df.describe().T[['mean', '50%', 'std']].rename(columns={'50%': 'median'})
print("\nSummary Statistics:")
print(summary_statistics)

# Create a correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

# Interpret correlation with Construction Cost
correlation_with_cost = df.corr()[dependent_variable].sort_values(ascending=False)
print("\nCorrelation of features with Construction_Cost:")
print(correlation_with_cost)
```
